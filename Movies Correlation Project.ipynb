```python
# Import libraries
 
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # This adjusts the configuration of the plots we will create

#Read in the data
df = pd.read_csv(r"C:\Users\USER\Desktop\excel proj\movies.csv")

#Lets look at the data
df.head()
```

```python
#Let's see if there is any missing data
# Let's create a for loop that loops through the data to see if there is anything missing

for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(missing*100)))

```

```python
# Check the data types for the columns

print(df.dtypes)
```

```python
# Check if there are any outliers. This helps us to  identify unusual observations, assess the data distribution, and understand relationships between variables. 

df.boxplot(column=['gross'])
```

```python
# Drop duplicates
df.drop_duplicates()
```

```python
# Sort our data a little bit
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# To check the relationship between gross and budget, we plot a scatter plot with a linear regression model
sns.regplot(x='gross', y='budget', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
# We then check the relationship between score and gross
sns.regplot(x='score', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
# Let's start looking at the correlation
```

```python
# Let's create a correlation matrix between all numeric columns

df.corr(method='pearson')
```

```python
# There's  a high correlation between budget and gross
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Values')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#We look at the correlation with company
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for movies')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#We fctorize the data again before correlation analysis 

correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

# Unstack the resulting matrix into a series
corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
# Let's sort corr_pairs using the quicksort algorithm

sorted_pairs = corr_pairs.sort_values(kind='quicksort')

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation of more than 0.5

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
